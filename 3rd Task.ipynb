```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
samplestore = pd.read_csv("SampleSuperstore.csv")
```

```python
samplestore.head()
```

```python
samplestore.describe()
```

```python
samplestore.dtypes
```

```python
samplestore.isnull().sum()
```

```python
corr = samplestore.corr()
sns.heatmap(corr, cmap="YlGnBu", annot=True)
plt.show()
```

```python
# Sales by Category
plt.figure(figsize=(12,6))
sns.barplot(x='Category', y='Sales', data=samplestore)
plt.show()
```

```python
# Profit by Category
plt.figure(figsize=(12,6))
sns.barplot(x='Category', y='Profit', data=samplestore)
plt.show()
```

```python
# Sales by Sub-Category
plt.figure(figsize=(16,8))
sns.barplot(x='Sub-Category', y='Sales', data=samplestore)
plt.show()
```

```python
# Profit by Sub-Category
plt.figure(figsize=(16,8))
sns.barplot(x='Sub-Category', y='Profit', data=samplestore)
plt.show()
```

```python
# Profit vs Sales
sns.jointplot(x='Sales', y='Profit', data=samplestore)
plt.show()
```
