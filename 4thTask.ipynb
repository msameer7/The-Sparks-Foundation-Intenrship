```python
import pandas as pd
```

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='latin-1')
```

```python
df.isnull().sum()
```

```python
# Display all column names
for column in df.columns:
    print(column)
```

```python
# Select relevant columns
columns = ['country_txt', 'region_txt', 'provstate', 'city', 'latitude', 'longitude']
df_geo = df[columns]
```

```python
# Count the number of terrorist incidents by country
country_counts = df_geo['country_txt'].value_counts()
```

```python
# Count the number of terrorist incidents by region
region_counts = df_geo['region_txt'].value_counts()
```

```python
# Count the number of terrorist incidents by state/province
state_counts = df_geo['provstate'].value_counts()
```

```python
# Count the number of terrorist incidents by city
city_counts = df_geo['city'].value_counts()
```

```python
# Print the top hot zones of terrorism
print("Top 10 countries with the highest number of terrorist incidents:")
print(country_counts.head(10))
```

```python
print("\nTop 10 regions with the highest number of terrorist incidents:")
print(region_counts.head(10))
```

```python
print("\nTop 10 states/provinces with the highest number of terrorist incidents:")
print(state_counts.head(10))
```

```python
print("\nTop 10 cities with the highest number of terrorist incidents:")
print(city_counts.head(10))
```

## DATA VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plotting histograms using Seaborn
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
sns.histplot(country_counts, bins=len(country_counts), kde=True)
plt.title('Top 10 Countries with the Highest Number of Terrorist Incidents')
```

```python
plt.subplot(2, 2, 2)
sns.histplot(region_counts, bins=len(region_counts), kde=True)
plt.title('Top 10 Regions with the Highest Number of Terrorist Incidents')
```

```python
plt.subplot(2, 2, 3)
sns.histplot(state_counts, bins=len(state_counts), kde=True)
plt.title('Top 10 States/Provinces with the Highest Number of Terrorist Incidents')
```

```python
# Plotting bar charts using Seaborn
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
sns.barplot(x=attack_counts.values, y=attack_counts.index, color='b')
plt.xlabel('Count')
plt.ylabel('Attack Type')
plt.title('Most Common Attack Types')
```

```python
plt.subplot(1, 2, 2)
sns.barplot(x=weapon_counts.values, y=weapon_counts.index, color='g')
plt.xlabel('Count')
plt.ylabel('Weapon Type')
plt.title('Most Common Weapon Types')

plt.tight_layout()
plt.show()
```

```python
# Plot the most common target types as a pie chart
plt.figure(figsize=(8, 6))
target_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Target Types')
plt.ylabel('')
```

```python
# Plot the most common target subtypes as a pie chart
plt.figure(figsize=(8, 6))
target_subtype_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Target Subtypes')
plt.ylabel('')
```

```python
# Plot the most common specific targets as a pie chart
plt.figure(figsize=(10, 8))
specific_target_counts.head(10).plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Specific Targets')
plt.ylabel('')

plt.show()
```
